```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('D:\Dân số\Data_visualization.xlsx')
df.head(5)
```

XEM DÂN SỐ QUA TỪNG NĂM

```python
group_population = df.groupby('Year').agg({'Population' : 'sum'})
group_population
```

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.grid(which = 'major', color = '#CCCCCC',linestyle = '--')
plt.plot( 'Population',data =group_population, marker = 'o' )
plt.title('Dân số từ 1955 - 2020')
plt.ylabel('Số lượng dân số (tỷ)')
plt.xlabel('Năm');
```

DÂN SỐ THEO VÙNG

```python
group_region = df.groupby('Region').sum()['Population']
group_region
```

```python
from random import randint
colors = ['r','b','g','yellow','orange','purple']
fig,ax = plt.subplots(figsize = (16,8))
plt.pie(group_region.values,
        colors = colors,
        autopct = '%1.1f%%',
        shadow = False,
        startangle=90);
plt.axis('equal');
plt.title('Dân số theo Châu Lục')
plt.legend(group_region.index,loc = 'upper left')
plt.show()
```

```python
df['Average_age'] = round(df['Average_age'])
df.head(5)
```

```python
group_age = df.groupby(['Region','Year']).agg({'Average_age':'mean'})
group_age
```

```python
df_subset = group_age.pivot_table (values = 'Average_age', index = 'Year',columns ='Region')
df_subset
```

```python
df_subset.plot(figsize = (12,10),grid = False, subplots = True,
               marker='o', markerfacecolor='blue', markersize=2, linewidth=2,title = 'TUỔI TRUNG BÌNH DỰA VÀO NĂM CỦA CÁC CHÂU LỤC');
```

 # TOP 5 NƯỚC  CÓ DÂN SỐ ĐÔNG NHẤT



```python
df.head(5)
```

```python
df_population_now = df.groupby("Country").agg({'Total_Population_2023' : 'max'}).sort_values(['Total_Population_2023'],ascending = False)
df_country_high = df_population_now[0:5]
df_country_high
```

```python
fig,ax = plt.subplots(figsize = (16,8))
colors = ['r','b','g','pink','orange']
plt.bar(x = df_country_high.index, height = df_country_high['Total_Population_2023'],color = colors)
plt.ylabel('Population (bilion)');
plt.xlabel('Country')
plt.title('TOP 5 QUỐC GIA CÓ DÂN SỐ LỚN NHẤT HIỆN NAY ( CẬP NHẬT THÁNG 3/2023 )')
labels = []
for i in range(len(df_country_high)):
    x = df_country_high['Total_Population_2023'][i]
    labels.append(x)
for i in range(len(df_country_high)):
    plt.text(x = df_country_high.index[i],
            y = df_country_high['Total_Population_2023'][i],
            s = labels[i] ,
            size = 10)
```

```python
df_country_top = df.pivot_table(values = 'Population',index = 'Year', columns = 'Country')[['Trung Quốc','Ấn Độ','Hoa Kỳ','Indonesia','Pakistan']]
df_country_top
```

```python
df_country_top.plot(figsize = (12,10),grid = False, subplots = True,
                    marker='o', markerfacecolor='blue', markersize=2, linewidth=2,title = 'TOP 5 DÂN SỐ QUA TỪNG NĂM TỪ 1955-2020');
```

```python
table_1 = df[['Country','Year','Population']]
table_1
```

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')

plt.plot('Year','Population',data = table_1[table_1["Country"] == "Trung Quốc"],marker='o', markerfacecolor='blue',
         markersize=3, color='skyblue', linewidth=3, label= "Trung Quốc")
plt.plot('Year','Population',data = table_1[table_1["Country"] == "Ấn Độ"],marker='o', markerfacecolor='blue',
         markersize=3, color='lightgreen', linewidth=3, label= "Ấn Độ")
plt.plot('Year','Population',data = table_1[table_1["Country"] == "Hoa Kỳ"],marker='o', markerfacecolor='blue',
         markersize=3, color='grey', linewidth=3, label= "Hoa Kỳ")
plt.plot('Year','Population',data = table_1[table_1["Country"] == "Indonesia"],marker='o', markerfacecolor='blue',
         markersize=3, color='red', linewidth=3, label= "Indonesia")
plt.plot('Year','Population',data = table_1[table_1["Country"] == "Pakistan"],marker='o', markerfacecolor='blue',
         markersize=3, color='cyan', linewidth=3, label= "Pakistan")
plt.legend()
plt.title('So sánh dân số từ năm 1955 - 2020 của top 5 quốc gia đông dân nhất thế giới');
```

```python
group_area = df.groupby('Country').agg({'Area' : 'max'})
group_population = df.groupby('Country').agg({'Population' : 'sum'})
table_2 = group_area.merge(group_population, left_on = 'Country', right_on = 'Country')
table_2
```

```python
import numpy as np
fig, ax = plt.subplots(figsize = (12,4),dpi = 80, facecolor = 'w', edgecolor = 'k')
np.random.seed(1223)
N = len(table_2)
colors = np.random.rand(N)
area = table_2.Area / 10000
plt.grid()
plt.scatter('Area','Population',data = table_2, c = colors ,s = area, cmap ='Reds',edgecolors = 'black', linewidths = .5)
plt.ylabel('Population')
plt.xlabel('Area')
plt.title('Mối liên hệ giữa diện tích và dân số các quốc gia');
plt.colorbar()
plt.show()
```

- Nga là nước có diện tích lớn nhất thế giới và gấp đôi so với nước thứ 2 là Trung Quốc nhưng dân số lại thua gấp rưỡi Trung Quốc => Bởi vì đa số lãnh thổ nước Nga là đồi núi và lạnh giá nên dân cứ khá thưa thớt
- Ấn Độ là nước có diện tích đứng thứ 7 thế giới nhưng dân số lại đứng thứ 2 thế giới => Ấn Độ đang có nguy cơ bùng nổ dân số và đáng báo động
- Trung Quốc mạnh cả về số lượng dân số và diện tích đất nước => Đang trên đường trở thành đất nước mạnh nhất thế giới vượt mặt cả Hoa Kỳ
